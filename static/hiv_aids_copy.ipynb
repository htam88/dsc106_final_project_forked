```python
import pandas as pd
import pycountry
import altair as alt
```

```python
aids_cases = pd.read_csv('aids_cases_final.csv').dropna()
aids_cases
```

```python
codes_and_coordinates = pd.read_csv('countries_codes_and_coordinates.csv')
codes_and_coordinates['Alpha-3 code'] = codes_and_coordinates['Alpha-3 code'].str.replace('"', '')
codes_and_coordinates = codes_and_coordinates.drop(columns=['Alpha-2 code', 'Numeric code'])
codes_and_coordinates['Alpha-3 code'] = codes_and_coordinates['Alpha-3 code'].str.strip()
codes_and_coordinates['Alpha-3 code'].unique()
```

```python
with_lat_long = aids_cases.merge(codes_and_coordinates, left_on='Country_Code', right_on='Alpha-3 code')
with_lat_long = with_lat_long.rename(columns = {'Country_y': 'Country'})
with_lat_long['Latitude (average)'] = with_lat_long['Latitude (average)'].str.replace('"', '')
with_lat_long['Longitude (average)'] = with_lat_long['Longitude (average)'].str.replace('"', '')
with_lat_long['Latitude (average)'] = with_lat_long['Latitude (average)'].astype(float)
with_lat_long['Longitude (average)'] = with_lat_long['Longitude (average)'].astype(float)
with_lat_long = with_lat_long.drop(columns=['Unnamed: 0'])
with_lat_long.to_csv('HIV_demographic_data.csv')
```
